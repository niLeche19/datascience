# Interest Calculator Program
This program starts with an initial money value of 1000 dollars and uses recursion to calculate compound intrest for the next 30 years.

In each recursive step, it uses the formula: $x_{2} = x_{1}*(1 + 0.01*\delta)$ to calculate the next value.

Where $x_{1}$ is the old interest value, $x_{2}$ is the new value and $\delta$ is the rate of change in percent.

And then produces a graph of the results.

```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import linspace
```

```python
dffvf = []
dffvs = []
def getni(x):
    return(x * (1 + 0.01 * 3.5))

def intnew(x, r):
    if r != 0:
        dffvs.append("$" + str(round(x)))
        dffvf.append(x)
        intnew(getni(x), (r - 1))
intnew(1000, 31)
```

```python
dff = pd.DataFrame()
dff['years'] = range(31)
dff['value_float'] = dffvf
dff['value_str'] = dffvs
```

```python
plt.plot(dff.years, dff.value_float)
plt.xlabel('Time elapsed (years)')
plt.ylabel('Sum of money (dollars)')

plt.show()
```
