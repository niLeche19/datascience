## Graphing electricity generation over time.

When looking for data for this notebook, my two criterion were: For the dataset to have between 300 and 500 points, and for it to have some sort of visible trend. While searching I found a sheet on electricity generated in the US over a period of about 40 years, which met both of my criterion. So I went with that and here is what I made.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default = 'warn'
```

```python
ec = pd.read_excel("http://bit.ly/2UO0FZL")
pyear = []
pkw = []
ls = 0
```

### First look at the data
Initially the data had a value for every month of the year. Because of things like temperature fluctuation per month, the graph had a very jagged and unorderly look.

```python
plt.plot(ec.time,ec.value)
plt.show()
```

## A little cleaning
As a solution for this I used two for loops to remove most of the data and just take the power generated in the months of January and February.


```python
for i in range(len(ec.time)):
    if i % 12 == 0 or i % 13 == 0:
        pyear.append(ec.time[i+1])

for i in range(len(ec.value)):
    if i % 12 == 0 or i % 13 == 0:
        pkw.append(ec.value[i+1])
```

```python
plt.plot(pyear,pkw)
plt.show()
```

This does not look stelar but it looks much better to me. A solution could be to remove the data from either January or February but I felt that that would be too little data and would be insufficient to draw conclusions.

So lets add labels and titles. 

```python
plt.plot(pyear,pkw)
plt.title('Electricity produced in the US over 42 years.')
plt.xlabel('Time elapsed (years)')
plt.ylabel('Electricity produced (kw x 1 billion)')
plt.show()
```
